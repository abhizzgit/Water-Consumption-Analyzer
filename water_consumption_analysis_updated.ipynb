# Water Consumption Pattern Analyzer

Includes hourly peak analysis.

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('water_consumption_hyderabad.csv')
df.head()
```

## Data Cleaning

```
df.dropna(inplace=True)
df['Date'] = pd.to_datetime(df['Date'])

df['Month'] = df['Date'].dt.month
df['Hour'] = df['Date'].dt.hour

df.head()
```

## Monthly Analysis

```
monthly_usage = df.groupby('Month')['Consumption'].sum()
monthly_usage.plot(marker='o', title='Monthly Water Consumption')
plt.show()
```

## Area-wise Analysis

```
area_usage = df.groupby('Area')['Consumption'].sum()
area_usage.plot(kind='bar', title='Water Consumption by Area')
plt.xticks(rotation=45)
plt.show()
```

## Hourly Peak Analysis

```
hourly_usage = df.groupby('Hour')['Consumption'].mean()

hourly_usage.plot(kind='bar', title='Average Hourly Water Consumption')
plt.xlabel('Hour of Day')
plt.ylabel('Average Consumption')
plt.show()

peak_hour = hourly_usage.idxmax()
print("Peak usage hour:", peak_hour)
```

## Heatmap

```
pivot = df.pivot_table(values='Consumption', index='Area', columns='Month', aggfunc='sum')

sns.heatmap(pivot, annot=True)
plt.title('Heatmap of Water Consumption')
plt.show()
```

## Insights

- Identified peak hourly usage  
- High consumption areas detected  
- Seasonal trends observed  

## Recommendations

- Manage supply during peak hours  
- Target high usage areas for conservation  
- Raise awareness during peak seasons
